### Extracting the top 30 follower's follower IDs

```python
# changing working directory

print("current working directory", os.getcwd())
change_dir('Follower_id')
print("current working directory", os.getcwd())
```

```python
def get_follower_id_userid(user_id):

    success, result = twitter_api('followers/ids.json?user_id=' + user_id)

    return result['ids']
```

```python
def store_follower_id_userid(user_id, filename):
    
    with open(filename, 'w+') as f:                                 # function overwrites not append
        for follower_id in get_follower_id_userid(user_id):
            f.write(str(user_id) + " " +str(follower_id) + "\n")
```

```python
# user_id list from top 30 followers
dfT30 = pd.read_csv("Top 30 Level 0-1 Followers.csv")

userid = []

for i in dfT30['id']:
    userid.append(str(i))

print(userid)
```

```python
# changing working directory

print("current working directory", os.getcwd())
change_dir('Raw (Level 1-2 Followers)')
print("current working directory", os.getcwd())
```

```python
# first 15

for i in range(0,15):
    print(userid[i])
    store_follower_id_userid(userid[i], 'follower_id_' + userid[i] + '.txt')
```

```python
# next 15

for i in range(15,30):
    print(userid[i])
    store_follower_id_userid(userid[i], 'follower_id_' + userid[i] + '.txt')
```

```python
dfT30F = pd.read_csv("follower_id_" + userid[0] + ".txt", sep=" ", header=None)

for i in range(1,30):
    # print(userid[i])
    temp2 = pd.read_csv("follower_id_" + userid[i] + ".txt", sep=" ", header=None)
    dfT30F = pd.concat([dfT30F, temp2], axis = 0)
```
